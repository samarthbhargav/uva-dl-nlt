```python
import matplotlib.pyplot as plt
import numpy as np
import codecs
import collections
%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = (15, 15)
```

```python
cat = {}
keys = []
with codecs.open("./data/reuters/cats.txt", "r", "utf-8") as reader:
    for line in reader:
        line = line.strip().split()
        _id = line[0]
        categories = line[1:]
        cat[_id] = categories
        keys.append(_id)
```

```python
def count(split):
    cat_counter = collections.defaultdict(int)
    for key, cats in cat.items():
        if key.startswith(split):
            for c in cats:
                cat_counter[c] += 1
    return cat_counter
test_counts = count("test")
train_counts = count("train")

train_counts = sorted(train_counts.items(), key=lambda _: _[1], reverse=True)
test_counts = sorted(test_counts.items(), key=lambda _: _[1], reverse=True)
```

```python
keys = [c[0] for c in train_counts]
values = [c[1] for c in train_counts]
plt.figure(figsize=(10, 20))
plt.barh(keys, values)
plt.title("Class Distribution for Train Split")
```

```python
keys = [c[0] for c in test_counts]
values = [c[1] for c in test_counts]
plt.figure(figsize=(10, 20))
plt.barh(keys, values)
plt.title("Class Distribution for Test Split")
```
